# Data Cleaning

```python
df["discounted_price"] = df["discounted_price"].astype(str)

df["discounted_price"] = df["discounted_price"].str.replace("₹","")

df["discounted_price"] = df["discounted_price"].str.replace(",","")

df["discounted_price"] = df["discounted_price"].astype(float)
```

- Convert the String column "discounted_price" to Float by removing "₹" and "," 

```python
df["actual_price"] = df["actual_price"].astype(str)

df["actual_price"] = df["actual_price"].str.replace("₹","")

df["actual_price"] = df["actual_price"].str.replace(",","")

df["actual_price"] = df["actual_price"].astype(float)
```

- Convert the String column "actual_price" to Float by removing "₹" and ","

```python
df['rating_count'] = df['rating_count'].astype(str)

df['rating_count'] = df['rating_count'].str.replace(",","")

df['rating_count'] = df['rating_count'].astype(float)
```

- Convert column 'rating_count' as Float with replacing ","

```python
df['rating'] = df['rating'].astype(str)

df['rating']=df['rating'].replace("|",4.1)

df['rating'] = df['rating'].astype(float)
```

- Convert column 'rating' as Float with replacing wrong value "|" to median value 4.1

```python
df['discount_percentage'] = df['discount_percentage'].astype(str)

df['discount_percentage'] = df['discount_percentage'].str.replace("%","")

df['discount_percentage'] = df['discount_percentage'].astype(int)
```

- Replace "%" symbol from 'discount_percentage' and Convert to Int

```python
df.describe()
```

```python
df["rating_count"] = df["rating_count"].fillna(1)
```

- Replaced 2 Null values with 1 because there is only 1 user for both the products

3. Variation in discount percentage – identifying products with the largest discounts

# Conclusion

From the Exploratory Data Analysis (EDA), several key insights emerge:

**1. Dominant Product Categories**

   - Electronics, Computers & Accessories, and Home & Kitchen dominate the dataset.
   - Electronics alone account for about **16.9%** of all products.

**2. Ratings Distribution**

   - Most products have ratings between **3.8 and 4.5**, indicating generally favorable reviews.
   - Extremely low or extremely high ratings are rare.

**3. Relationship Between Price and Discounts**

   - Strong positive correlation **0.96** between actual price and discounted price -- **expensive products remain expensive even after discounts**.
   - Discount percentage has a slight negative correlation **-0.24** with price -- **higher-priced items often receive smaller relative discounts**.
   - Slight negative relationship between rating and discount percentage, suggesting heavy discounts do not necessarily mean better ratings.

**4. Price Distribution**

   - Most products are priced below **₹20,000**.
   - Distribution is heavily right-skewed due to a few high-end or luxury products.

**5. Discounts vs Ratings**

   - Highly rated products do not always have the biggest discounts.
   - Moderate discounts are more common for well-rated items.
   - Some outliers exist with extreme discounts or unusually high prices regardless of rating.

**6. Consumer Behavior Implications**

   - High discounts alone are not enough to guarantee high ratings -- **quality matters more**.
   - In competitive categories like Electronics, differentiators such as quality assurance and brand reputation are key.
