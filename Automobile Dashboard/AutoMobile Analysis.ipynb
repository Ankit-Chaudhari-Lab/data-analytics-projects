# Data Cleaning

```python
df = df.drop(columns = "Unnamed: 0")
```

Column "Unnamed: 0" has been dropped

```python
df[df["stroke"].isnull()].index
```

```python
df["stroke"] = df["stroke"].fillna(df["stroke"].mean())
```

- Solution :- Missing Stroke values has been treated with mean values

```python
df[df["horsepower-binned"].isnull()].index
```

```python
df["horsepower-binned"] = df["horsepower-binned"].fillna(df["horsepower-binned"].mode()[0])
```

- Solution :- Missing horsepower-binned value have been treated with mode value

```python
df.isnull().sum()
```

- Observation :- There is no missing values are available in our Data

```python
df["peak-rpm"].astype(int)
```

```python
df["price"].astype(int)
```

- Solution :- "peak-rpm" and "price" column has been converdet to integer(int)

```python
df["diesel"] = df["diesel"].replace({0:"No", 1:"Yes"})
```

```python
df["gas"] = df["gas"].replace({0:"No", 1:"Yes"})
```

```python
df = df.rename(columns = {"make":"brand"})
```

- Solution :- Remane make column to brand

```python
df.to_excel("Cleaned_Automobile_Dataset.xlsx")
```
